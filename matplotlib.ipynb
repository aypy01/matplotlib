##  Basic Line Plotting

```python
#Importing of essential libraries
import numpy as np
import matplotlib.pyplot as plt
```

### Plotting y = 2x

```python
# Define x and y values for a simple linear relationship: y = 2x
x = [0,1,2,3,4,5,6,7,8]
y = [0,2,4,6,8,10,12,14,16]

# Plot the line
plt.plot(x, y, label='y = 2x', color='blue', marker='o')

# Add labels and title
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Simple Line Plot: y = 2x")
plt.legend()
plt.grid(True)
plt.show()
```

###  Plotting Quadratic Function with Styles

```python
# Create x values using numpy arange
x2 = np.arange(0, 6, 0.5)

# Plot first segment of x^2 in red
plt.plot(x2[:5], x2[:5]**2, color='red', label='xÂ² (solid)', linewidth=2)

# Plot remaining segment in dotted orange
plt.plot(x2[4:], x2[4:]**2, linestyle='dotted', color='orange', linewidth=3, label='xÂ² (projection)')

# Final touches
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Styled Quadratic Plot")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Bar Chart

labels=['A','B','C','D','E']
values=[1,2,3,4,5]

plt.figure(figsize=(5,5),dpi=150) #Figure is an intial part and there for showing not saving

bars=plt.bar(labels,values)

patterns=['/','*','o','|','-',]

for bar in bars:
  bar.set_hatch(patterns.pop(-1))# Pop (-1) select the patter from pattern list and the prints on the bar from last to first (-1)

plt.savefig('bar_chart.png',dpi=300) #Now this will save in this specifaction of that chart

plt.show()
```

### Pie Chart Example

```python
# Preprocessing the 'Weight' column
# Assuming 'Weight' is in the format '150lbs', '165lbs', etc.
# Ensure Weight is a string before stripping 'lbs'
player['Weight'] = player['Weight'].astype(str).str.replace('lbs', '').astype(float)


# Initialization
plt.figure(figsize=(4, 4), dpi=200)
plt.style.use('ggplot')
plt.title("Weight of Professional FIFA Players (in lbs)")

# Count the categories
light = player[player['Weight'] < 125].shape[0]
light_medium = player[(player['Weight'] >= 125) & (player['Weight'] < 150)].shape[0]
medium = player[(player['Weight'] >= 150) & (player['Weight'] < 175)].shape[0]
medium_heavy = player[(player['Weight'] >= 175) & (player['Weight'] < 200)].shape[0]
heavy = player[player['Weight'] >= 200].shape[0]

# Labels and data
weights = [light, light_medium, medium, medium_heavy, heavy]
labels = ["<125", "125â€“149", "150â€“174", "175â€“199", "200+"]
colors = ['#08ea47', '#d40e81', '#8c58aa', '#d40ec9', '#d4a30e']

#Explode as the graph is so conjested
explode=(0.5,0.2,0.1,0,0.4)
# Plot
plt.pie(weights, labels=labels, colors=colors, autopct='%0.2f%%',explode=explode,pctdistance=0.5) #PCT distance adjust the pct offset
plt.show()
```
